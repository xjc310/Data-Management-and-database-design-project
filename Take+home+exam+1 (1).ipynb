## 18.Why create temporary tables?

######  Temporary table allows users to store a temporary result set, which users can reuse several times in a single session. A temporary table is very handy when it is impossible or expensive to query data that requires a single SELECT statement with JOIN clauses. In this case, you can use a temporary table to store the immediate result and use another query to process it.

### 28 Explain to an eight year old (i.e. your professor) what are the first  three Normal Forms. (10 points)


#### 1st normal form: elminate duplicative columns from the same table. Besides, create seperate tables for each group of related table  data and identify each row with a unique column. 2nd normal form: Meet the requirement of the 1st normal form. Besides,  create relationships between these new tables and their predecessors through the use of foreign keys. 3nd normal form: Already meet the requirements of both 1NF and 2NF. Remove columns that are not fully dependent upon the primary key.
